```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

from selenium.common.exceptions import NoSuchElementException

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
def wait_return_text (driver, loc):
    try:
        return WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.CSS_SELECTOR, loc))).text
    except:
        return False

def nextpage(driver):
    try:
        element = WebDriverWait(driver, 3).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "#results > ul.sc-kGdDhL.kJFtWa > li:nth-child(12) > button > svg"))
        )
        element.click()
    except Exception as e:
        return False
```

```python
url_main = "https://www.kaggle.com/search?q=Customer+Churn+in%3Adatasets"
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get(url_main)
```

```python
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

driver.get(url_main)
original_window = driver.current_window_handle
df = []
while 1:
    i = 0
    while 1:
        #ÏÉàÏ∞ΩÏó¥Í∏∞
        try:
            i +=1
            element = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.CSS_SELECTOR, f"#results > ul.MuiList-root.km-list.css-1uzmcsd > li:nth-child({i}) > div > a > div > div.sc-bjhZkk.jWyUHl > div")))
            ActionChains(driver)\
                .key_down(Keys.CONTROL)\
                .click(element)\
                .key_up(Keys.CONTROL)\
                .perform()
            time.sleep(1)  # ÏÉà Ï∞ΩÏù¥ Ïó¥Î¶¥ ÎïåÍπåÏßÄ Ïû†Ïãú ÎåÄÍ∏∞
            driver.switch_to.window(driver.window_handles[-1])
        except:
            break

        time.sleep(2)
        try:     #site-content > div:nth-child(2) > div > div.sc-iQQCXo.gSQcTP > div > div.sc-iqroHm.kuxaKs > div.sc-kYuORB.jKlOrq > div > div > div.sc-eGrtvL.dXLXG > div > div.sc-fnWRuw.iQZdfc > div > div.sc-bsjWWi.eXCtlv > div > div.MuiTabs-scroller.MuiTabs-hideScrollbar.MuiTabs-scrollableX.css-12qnib > div > button:nth-child(3) > span.sc-dKKIkQ.kcQzVn
            t = "#site-content > div:nth-child(2) > div > div.sc-iQQCXo.gSQcTP > div > div.sc-iqroHm.kuxaKs > div.sc-kYuORB.jKlOrq > div > div > div.sc-eGrtvL.dXLXG > div > div.sc-fnWRuw.iQZdfc > div > div.sc-bsjWWi.eXCtlv > div > div.MuiTabs-scroller.MuiTabs-hideScrollbar.MuiTabs-scrollableX.css-12qnib > div > button:nth-child(3) > span.sc-dKKIkQ.kcQzVn"
            WebDriverWait(driver, 1).until(EC.element_to_be_clickable((By.CSS_SELECTOR, t))).click()
        except:
            tmp_link = driver.current_url
            row = [False, False, False, False, tmp_link]
            df.append(row)
            driver.close()
            driver.switch_to.window(original_window)
            continue
        time.sleep(2)
        name= wait_return_text(driver, "#site-content > div:nth-child(2) > div > div.sc-iQQCXo.gSQcTP > div > div.sc-yVOhj.kmmEya > div.sc-jxyWrI.JAEoS.sc-hocAEa.kXCOub > div.sc-lcsTPt.givKIP > h1")
        columns = wait_return_text(driver, "#site-content > div:nth-child(2) > div > div.sc-iQQCXo.gSQcTP > div > div.sc-iqroHm.kuxaKs > div.sc-kYuORB.jKlOrq > div > div > div.sc-eGrtvL.dXLXG > div > div.sc-fnWRuw.iQZdfc > div > div.sc-ftpwRR.jvLrRV > button > p")
        if columns:
            columns = columns.split(" ")[2]
        usability = wait_return_text(driver, "#site-content > div:nth-child(2) > div > div.sc-iQQCXo.gSQcTP > div > div.sc-iqroHm.kuxaKs > div.sc-jckpuY.iNrxIt > div.sc-mXAHQ.duwXti > p:nth-child(2) > span")
        rows = wait_return_text(driver, "#site-content > div:nth-child(2) > div > div.sc-iQQCXo.gSQcTP > div > div.sc-iqroHm.kuxaKs > div.sc-kYuORB.jKlOrq > div > div > div.sc-eGrtvL.dXLXG > div > div.sc-jMwPYW.loWIXM > div > div:nth-child(1) > div.sc-iHntzg.bGtlwD > div > div > div:nth-child(2) > span.sc-gGarWV.sc-kYoIAe.ekfEGc.GhHej")
        tmp_link = driver.current_url
        row = [name, rows, columns, usability, tmp_link]
        if not row in df:
            df.append(row)
            print(i, row)
        driver.close()
        driver.switch_to.window(original_window)
        time.sleep(1)
    try:
        time.sleep(1)
        element = WebDriverWait(driver, 1).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "#results > ul.sc-kGdDhL.kJFtWa > li:nth-child(12) > button > svg"))
        )
        element.click()
        time.sleep(1)
    except:
        break
```
