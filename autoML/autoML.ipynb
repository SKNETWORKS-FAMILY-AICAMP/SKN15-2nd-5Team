```python
from pycaret.classification import setup, compare_models, pull
import pandas as pd
```

```python
df = pd.read_csv("../dataSet/final_dataset.csv", encoding='utf-8')
df.shape
```

```python
s = setup(data=df, target='churn', session_id=42, fix_imbalance=True)
compare_models()
results = pull()
print(results[['Model', 'Recall']].sort_values('Recall', ascending=False))
```
